### Input

```python
df = pd.read_csv(r".\train\train.csv")
column_name_list = df.columns.tolist()
```

```python
# 3x3のsubplotを作成
fig, axs = plt.subplots(3, 3, figsize=(15, 15))

# ヒストグラム
sns.histplot(df['acousticness'], bins=10, kde=True, ax=axs[0, 0])
axs[0, 0].set_title('Acousticness Histogram')

# 箱ひげ図
sns.boxplot(x=df['danceability'], ax=axs[0, 1])
axs[0, 1].set_title('Danceability Boxplot')

# 散布図
sns.scatterplot(x='energy', y='danceability', data=df, ax=axs[0, 2])
axs[0, 2].set_title('Energy vs Danceability Scatter Plot')

# 相関行列とヒートマップ
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True, ax=axs[1, 0])
axs[1, 0].set_title('Correlation Heatmap')

# Pairplot（時間がかかる場合があるため、必要な特徴だけ選択することも可能）
# sns.pairplot(df[['acousticness', 'danceability', 'energy']], ax=axs[1, 1])
# axs[1, 1].set_title('Selected Features Pairplot')

# KDE Plot
sns.kdeplot(data=df[['acousticness', 'danceability', 'energy']], ax=axs[1, 2])
axs[1, 2].set_title('KDE Plot for Selected Features')

# Violin Plot
sns.violinplot(x='instrumentalness', data=df, ax=axs[2, 0])
axs[2, 0].set_title('Instrumentalness Violin Plot')

plt.tight_layout()
plt.show()
```
